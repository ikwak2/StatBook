## 정규분포 확률 구하기 
<예 6-3>에서 확률변수 X가 X~N(20,25)일 때, X가 15에서 30 사이일 확률을 R을 이용하여 구한다.

### 확률의 계산
pnorm 명령문을 사용하여 정규분포 확률값을 구한다. 5장과 마찬가지로 pnorm은 normal 분포의 줄임말로 norm과 접두어 p가 붙은 명령어이다.

```r
# X~N(20,25) 일 때, P(15 < X < 30) 의 확률 
pnorm(30, mean=20, sd=5)-pnorm(15, mean=20, sd=5)
#정규분포의 누적확률값을 나타내는 pnorm 명령문을 사용한다.
#pnorm(x, mean=평균, sd=표준편자)
```

### 정규분포 확률곡선 그리기
 plot 명령문을 사용하여 정규분포 그래프를 그린다.

```r
x<-seq(1, 40)
# x 데이터셋에 1에서 40까지의 정수를 대입한다.
y<-dnorm(x, mean=20, sd=5)
# y 데이터셋에서는 dnorm을 이용하여 x값에 상응하는 평균이 20이고 표준편차가 5인 정규분포 확률값을 산출한다. 
plot(x, y, type="l", main="Normal Distribution")
# plot(x축 데이터, y축 데이터, type=선의 종류, main=“큰제목”)
abline(v=20, col="red", lty=2)
# abline(v=수직선 위치 지정, col=선의 색깔, lty=선의 종류)
# abline은 좌표에 직선을 그리는 명령어로, v(수직선), h(수평선)를 이용하여 아래와 같이 보조선을 추가할 수 있다.
```

## 지수분포 확률 구하기 
<예 6-8>에서 전화가 한 번 걸려온 후에 다음 전화가 걸려올 때까지 걸린 시간이 5분 이내일 확률과 2분 이상일 확률은 구한다.

### 확률의 계산
pexp 명령문을 사용하여 지수분포 확률값을 구한다. pexp는 지수분포(exponential 분포)의 줄임말로 exp와 접두어 p가 붙은 명령어이다.

```r
# 지수분포의 누적확률값을 나타내는 pexp 명령문을 사용한다.
# pexp(x, rate=단위시간당 평균 사건 발생 비율)
pexp(5, rate=0.5)  #X<=5일 확률
```

```r
1-pexp(2, rate=0.5)  #X>=2일 확률
```

### 지수분포 확률곡선 그리기
plot 명령문을 사용하여 지수분포 그래프를 그린다.

```r
x<-seq(1, 20)
# x 데이터셋에 1에서 20까지의 정수를 대입한다.
y<-dexp(x, 0.5)
# y 데이터셋에서는 dexp를 이용하여 x값에 상응하는 평균 발생 값이 0.5인 지수분포의 확률값을 산출한다. 
plot(x, y, type="l", main="Exponential Distribution") 
```
