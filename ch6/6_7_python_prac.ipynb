## 정규분포 확률 구하기 
<예 6-3>에서 확률변수 X가 X~N(20,25)일 때, X가 15에서 30 사이일 확률을 R을 이용하여 구한다.

### 확률의 계산
norm.cdf 명령문을 사용하여 정규분포 확률값을 구한다. 

```python
# X~N(20,25) 일 때, P(15 < X < 30) 의 확률 
from scipy.stats import norm
norm.cdf(30, loc=20, scale=5) - norm.cdf(15, loc=20, scale=5)

#정규분포의 누적확률값을 나타내는 norm.cdf 명령문을 사용한다.
#norm.cdf(x, loc=평균, scale=표준편자)
```

### 정규분포 확률곡선 그리기
 plot 명령문을 사용하여 정규분포 그래프를 그린다.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.arange(1, 41) # x 데이터셋에 1에서 40까지의 정수를 대입한다.
y = norm.pdf(x, loc=20, scale=5) # y 데이터셋에서는 norm.pdf를 이용하여 x값에 상응하는 평균이 20이고 표준편차가 5인 정규분포 확률값을 산출한다. 

plt.plot(x, y, 'b-', linewidth=2) # .plot(x축 데이터, y축 데이터, ... )
plt.axvline(x=20, color='red', linestyle='--') # .axvline 은 세로 선을 그려주는 함수
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

## 지수분포 확률 구하기 
<예 6-8>에서 전화가 한 번 걸려온 후에 다음 전화가 걸려올 때까지 걸린 시간이 5분 이내일 확률과 2분 이상일 확률은 구한다.

### 확률의 계산
expon.cdf 명령문을 사용하여 지수분포 확률값을 구한다. 

```python
from scipy.stats import expon

expon.cdf(5, scale=1/0.5) #X<=5일 확률
# 지수분포의 누적확률값을 나타내는 expon.cdf 명령문을 사용한다.
# expon.cdf(x, scale=이벤트 1회 발생하는 평균시간)
```

```python
1-expon.cdf(2, scale=1/0.5)  #X>=2일 확률
```

### 지수분포 확률곡선 그리기
plt.plot 명령문을 사용하여 지수분포 그래프를 그린다.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

x = np.arange(1, 21) # x 데이터셋에 1에서 20까지의 정수를 대입한다.
y = expon.pdf(x, scale=1/0.5)  #  y 데이터셋에서는 expon.pdf를 이용하여 x값에 상응하는 평균 발생 값이 0.5인 지수분포의 확률값을 산출한다. 

plt.plot(x, y, 'b-', linewidth=2)
plt.title('Exponential Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```
